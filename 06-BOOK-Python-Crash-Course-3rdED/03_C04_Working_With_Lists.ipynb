#### Looping Through an Entire List


```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians:
    print(magician.title())
```

```python
magicians = ['alice', 'david', 'carolina']

for index, magician in enumerate(magicians):
    print(f'Item {index} : {magician.title()}')

# enumerate() gives you both the index and the value at the same time, which is perfect for this situation.
```

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")
```

```python
magicians = ['alice', 'david', 'carolina']

for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")
    print(f"I can't wait to see your next trick, {magician.title()}.\n")

print("Thank you, everyone. That was a great magic show!")
```

#### Using **range()** to Make a List of Numbers


```python
numbers = list(range(1, 6))
print(numbers)
```

```python
even_numbers = list(range(2, 11, 2))
print(even_numbers)
```

```python
squares = []

for value in range(1, 11):
    square = value ** 2
    squares.append(square)
    
print(squares)
```

```python
squares = []

for value in range(1,11):
    squares.append(value**2)
    
print(squares)
```

#### List Comprehensions


```python
# Create a list called 'squares' using a list comprehension
# The expression 'value**2' means "square the number"
# 'for value in range(1, 11)' generates numbers from 1 through 10
# For each number in that range, Python calculates value**2 and stores it in the list
squares = [value**2 for value in range(1, 11)]

# Print the resulting list of squares
print(squares)

# This is a compact way to generate lists compared to writing a loop with .append()
```

```python
cubes = [value**3 for value in range (1, 11)]

print(cubes)
```

#### Looping Through a Slice


```python
players = ['charles', 'martina', 'michael', 'florence', 'eli']

print("Here are the first three players on my team:")

for player in players[:3]:
    print(player.title())
```

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli']

print("Here are the last three players on my team:")

for player in players[:1:-1]:
    print(player.title())
```

#### Copying a List


```python
my_foods = ['pizza', 'falafel', 'carrot cake']

friend_foods = my_foods[:]

print("My favorite foods are:")
print(my_foods)

print("\nMy friend's favorite foods are:")
print(friend_foods)
```

```python
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]

my_foods.append('cannoli')
friend_foods.append('ice cream')

print("My favorite foods are:")
print(my_foods)

print("\nMy friend's favorite foods are:")
print(friend_foods)
```

```python
my_foods = ['pizza', 'falafel', 'carrot cake']

# This doesn't work:
friend_foods = my_foods

my_foods.append('cannoli')
friend_foods.append('ice cream')

print("My favorite foods are:")
print(my_foods)

print("\nMy friend's favorite foods are:")
print(friend_foods)
```
