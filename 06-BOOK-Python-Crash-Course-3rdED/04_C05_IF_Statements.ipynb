#### A Simple Example


```python
cars = ['audi', 'bmw', 'subaru', 'toyota']
for car in cars:
    if car == 'bmw':
        print(car.upper())
    else:
        print(car.title())
```

#### Checking for Equality


```python
car = 'bmw'

car == 'bmw'
```

```python
car = 'audi'

car == 'bmw'
```

#### Ignoring Case When Checking for Equality


```python
car = 'Audi'

car == 'audi'
```

```python
car = 'Audi'

car.lower() == 'audi'
```

```python
car = 'Audi'

car.lower() == 'audi'

# print(car) 
```

#### Checking for Inequality


```python
requested_topping = 'mushrooms'

if requested_topping != 'anchovies':
    print("Hold the anchovies!")
```
