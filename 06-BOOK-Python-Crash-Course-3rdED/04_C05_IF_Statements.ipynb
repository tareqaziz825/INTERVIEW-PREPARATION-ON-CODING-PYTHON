#### TRY IT YOURSELF (5-1. Conditional Tests)

```python
car = 'subaru'
print("Is car == 'subaru'? I predict True.")
print(car == 'subaru')

print("\nIs car == 'audi'? I predict False.")
print(car == 'audi')

age = 20
print("\nIs age >= 18? I predict True.")
print(age >= 18)

print("\nIs age < 18? I predict False.")
print(age < 18)

fruit = 'apple'
print("\nIs fruit.lower() == 'apple'? I predict True.")
print(fruit.lower() == 'apple')

print("\nIs fruit == 'Apple'? I predict False.")
print(fruit == 'Apple')

numbers = [1, 2, 3, 4, 5]
print("\nIs 3 in numbers? I predict True.")
print(3 in numbers)

print("\nIs 10 in numbers? I predict False.")
print(10 in numbers)

requested_toppings = ['mushrooms', 'onions', 'pineapple']
print("\nIs 'mushrooms' in requested_toppings? I predict True.")
print('mushrooms' in requested_toppings)

print("\nIs 'pepperoni' in requested_toppings? I predict False.")
print('pepperoni' in requested_toppings)
```

```python
car = 'subaru'
age = 20
fruit = 'apple'
numbers = [1, 2, 3, 4, 5]
requested_toppings = ['mushrooms', 'onions', 'pineapple']

# AND operator
print("Is car == 'subaru' and age > 18? I predict True.")
print(car == 'subaru' and age > 18)

print("\nIs car == 'audi' and age > 18? I predict False.")
print(car == 'audi' and age > 18)

# OR operator
print("\nIs car == 'subaru' or fruit == 'banana'? I predict True.")
print(car == 'subaru' or fruit == 'banana')

print("\nIs car == 'audi' or fruit == 'banana'? I predict False.")
print(car == 'audi' or fruit == 'banana')

# NOT operator
print("\nIs not (age < 18)? I predict True.")
print(not (age < 18))

print("\nIs not (fruit == 'apple')? I predict False.")
print(not (fruit == 'apple'))

# Combining membership with logical operators
print("\nIs 'mushrooms' in requested_toppings and 'pepperoni' not in requested_toppings? I predict True.")
print('mushrooms' in requested_toppings and 'pepperoni' not in requested_toppings)

print("\nIs 3 in numbers or 10 in numbers? I predict True.")
print(3 in numbers or 10 in numbers)

print("\nIs 10 not in numbers? I predict True.")
print(10 not in numbers)

print("\nIs 'onions' in requested_toppings and car == 'audi'? I predict False.")
print('onions' in requested_toppings and car == 'audi')
```

#### TRY IT YOURSELF (5-2. More Conditional Tests)
