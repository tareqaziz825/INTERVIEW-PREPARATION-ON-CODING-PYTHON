    merged = [intervals[0]]

def binary_search(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

### Question: Two Sum

Given an array of integers `nums` and an integer `target`, return the **indices** of the two numbers such that they add up to the target.

**Brief Explanation:**
You need to find two distinct positions in the list where the values at those positions sum exactly to the target value. You cannot use the same element twice, and you can assume each input has exactly one solution.


###  How the Solution Works

This solution uses a **Hash Map** (Python dictionary) to store numbers we have already "seen" as we walk through the list.

* **Initialization:** We create an empty dictionary `seen` where the **key** is the number and the **value** is its index.
* **The Complement Check:** For every number (`num`) in the list, we calculate the `complement` (). This is the specific value needed to reach the target.
* **The Lookup:**
* If the `complement` is already in our `seen` dictionary, it means we found the pair! We return the index of that complement and our current index `i`.
* If itâ€™s not there, we store the current `num` and its index in the dictionary and move to the next item.


### Complexity Analysis

#### Time Complexity: 

* **Iteration:** We traverse the list containing  elements exactly once.
* **Lookup:** Checking if the `complement` exists in a hash map (dictionary) takes, on average,  time.
* **Total:** .

#### Space Complexity: 

* In the worst-case scenario (e.g., the two numbers are at the very end of the list), we will store  elements in the dictionary, resulting in linear space usage.


````python
def two_sum(nums, target):
    seen = {}
    # Manually iterating through indices using range()
    for i in range(len(nums)):
        num = nums[i]
        complement = target - num
        
        if complement in seen:
            return [seen[complement], i]
        
        seen[num] = i
    return []

# --- Driver Code ---
if __name__ == "__main__":
    # Define test cases: (list of numbers, target sum)
    test_nums = [2, 7, 11, 15]
    test_target = 22

    # Call the function and store the result
    result = two_sum(test_nums, test_target)

    # Output the findings
    if result:
        print(f"Target {test_target} found!")
        print(f"Indices: {result}")
        print(f"Values: {test_nums[result[0]]} + {test_nums[result[1]]} = {test_target}")
    else:
        print("No two numbers sum up to the target.")
````

### Question: Longest Substring Without Repeating Characters

Given a string `s`, find the length of the **longest substring** that contains no duplicate characters.

**Brief Explanation:**
A "substring" must be a contiguous sequence of characters. The goal is to find the widest "window" within the string where every character inside that window is unique.


### How the Solution Works

This solution uses the **Sliding Window** technique with two pointers (`left` and `right`) and a `set` for instant lookups.

* **The Window:** The range between `left` and `right` represents our current substring.
* **Expanding (Right Pointer):** We iterate through the string using the `right` pointer. For every new character, we check if itâ€™s already in our `char_set`.
* **Shrinking (Left Pointer):** If we encounter a character that is **already** in the set (a duplicate), we have a conflict. We move the `left` pointer forward and remove characters from the set until the duplicate is gone.
* **Update Max:** After ensuring the window is "clean" (no duplicates), we add the current character to the set and calculate the window size (). If this is larger than our previous `max_len`, we update it.


### Complexity Analysis

#### Time Complexity: 

* **Traversal:** Although there is a `while` loop inside a `for` loop, each character is visited by the `right` pointer once and the `left` pointer at most once.
* **Operations:** Set operations (`add`, `remove`, `in`) are  on average.
* **Total:** In the worst case, each pointer travels the length of the string , resulting in  operations, which simplifies to .

#### Space Complexity: 

* The size of the `char_set` depends on the size of the string () and the size of the character alphabet (). If the string is very long but only uses 26 lowercase letters, the set will never exceed 26 elements.

````python
def length_of_longest_substring(s):
    char_set = set()
    left = 0
    max_len = 0

    for right in range(len(s)):
        # If we hit a duplicate, shrink the window from the left
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        # Calculate window size: (right index - left index + 1)
        max_len = max(max_len, right - left + 1)

    return max_len

# --- Driver Code ---
if __name__ == "__main__":
    # Test cases: (Input string, Expected result)
    test_cases = [
        ("abcabcbb", 3),  # "abc"
        ("bbbbb", 1),     # "b"
        ("pwwkew", 3),    # "wke"
        ("", 0),          # Empty string
        (" ", 1),         # Single space
        ("au", 2)         # Two distinct
    ]

    print(f"{'Input String':<15} | {'Result':<10} | {'Status'}")
    print("-" * 40)

    for text, expected in test_cases:
        result = length_of_longest_substring(text)
        status = "âœ… Pass" if result == expected else "âŒ Fail"
        print(f"'{text}':<15 | {result:<10} | {status}")
````

### Question: Maximum Subarray Sum

Given an integer array `nums`, find the **contiguous** subarray (containing at least one number) which has the largest sum and return its sum.

**Brief Explanation:**
A "subarray" is a slice of the original array. This problem asks you to find a "hot spot" within the array where the numbers added together give you the highest possible value, even if some of those numbers are negative.

---

### How the Solution Works

This solution uses **Kadaneâ€™s Algorithm**, a classic dynamic programming approach that makes a local vs. global decision at every step.

* **Initial State:** We start by setting both `current` (the best sum ending at the current position) and `max_sum` (the overall best sum found so far) to the first element of the array.
* **The Choice:** As we loop through the rest of the numbers (`nums[1:]`), we ask one question at each index:
* "Is it better to add this number to my existing streak, or should I just start a new streak starting with this number?"
* `current = max(num, current + num)`


* **The Update:** If our `current` streak is better than our historical `max_sum`, we update `max_sum`.

Essentially, if a prefix sum becomes negative, Kadaneâ€™s tells us to "drop" it and start fresh, because a negative sum will only drag down the value of any future elements.

---

### Complexity Analysis

#### Time Complexity: 

* **Traversal:** We iterate through the array exactly once ( elements).
* **Operations:** Inside the loop, we perform simple constant-time arithmetic and comparison operations.
* **Total:** .

#### Space Complexity: 

* **Efficiency:** We only use two variables (`current` and `max_sum`) regardless of how large the input array is. We do not use any extra data structures like lists or sets.

---

````python
def max_subarray(nums):
    # Handle empty list edge case
    if not nums:
        return 0
        
    current = max_sum = nums[0]

    for num in nums[1:]:
        # Decide: Start over at 'num' or add 'num' to the existing 'current' sum
        current = max(num, current + num)
        # Record the highest 'current' we've ever seen
        max_sum = max(max_sum, current)

    return max_sum

# --- Driver Code ---
if __name__ == "__main__":
    test_cases = [
        ([-2, 1, -3, 4, -1, 2, 1, -5, 4], 6),  # Subarray: [4, -1, 2, 1]
        ([1], 1),                              # Single element
        ([5, 4, -1, 7, 8], 23),                # All mostly positive
        ([-8, -3, -6, -2, -5, -4], -2)         # All negative (picks the largest single element)
    ]

    print(f"{'Input Array':<30} | {'Max Sum':<10} | {'Status'}")
    print("-" * 55)

    for nums, expected in test_cases:
        result = max_subarray(nums)
        status = "âœ… Pass" if result == expected else "âŒ Fail"
        print(f"{str(nums):<30} | {result:<10} | {status}")
````
