### 1Ô∏è‚É£ Reverse Words in a String

**Problem:**
Reverse the order of words in a string. Words are separated by spaces.

**Tricky Cases:**

- Multiple spaces between words
- Leading or trailing spaces
- Empty string

**Example:**

```
Input: "  hello   world  "
Output: "world hello"
```

**Hint:**
Trim spaces and split intelligently. Don‚Äôt just reverse string blindly.

# üß† Samsung Edge-Case Thinking Tips

1. **Always ask about constraints:** Can input be empty? negative? large? duplicate?
2. **Think about boundaries:** 0, 1, n, max int, min int.
3. **Test manually:** Before submitting, imagine small arrays like `[0]`, `[1,1]`, `[]`.
4. **Brute force first in mind:** Helps identify tricky cases before optimizing.

### 1Ô∏è‚É£ Reverse Words in a String

```python
def reverse_words(s):
    # Split by whitespace and remove empty strings
    words = s.strip().split()
    words = [word for word in words if word]
    # Reverse the list
    return ' '.join(words[::-1])

# Edge Cases
print(reverse_words("  hello   world  "))  # "world hello"
print(reverse_words(""))  # ""
print(reverse_words("single"))  # "single"
```

**Time Complexity:** O(n)
**Edge Case Handling:** Multiple spaces, leading/trailing spaces, empty string.

### 2Ô∏è‚É£ Two Sum With Duplicates

```python
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Edge Cases
print(two_sum([3,3], 6))  # [0,1]
print(two_sum([1,-1,0], 0))  # [0,1]
print(two_sum([1,2,3], 7))  # []
```

**Time Complexity:** O(n)
Handles duplicates and negative numbers.

### 3Ô∏è‚É£ Palindrome With Special Characters

```python
import re

def is_palindrome(s):
    cleaned = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return cleaned == cleaned[::-1]

# Edge Cases
print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome(""))  # True
print(is_palindrome("ab"))  # False
```

**Time Complexity:** O(n)
Handles spaces, punctuation, case sensitivity, empty string.

### 4Ô∏è‚É£ Maximum Product Subarray

```python
def max_product(nums):
    if not nums:
        return 0
    
    max_prod = min_prod = result = nums[0]
    
    for n in nums[1:]:
        if n < 0:
            max_prod, min_prod = min_prod, max_prod
        
        max_prod = max(n, n*max_prod)
        min_prod = min(n, n*min_prod)
        
        result = max(result, max_prod)
    
    return result

# Edge Cases
print(max_product([2,3,-2,4]))  # 6
print(max_product([-2,0,-1]))   # 0
print(max_product([-2,-3,0,-2])) # 6
```

**Time Complexity:** O(n)
Handles negatives and zeros.

### 5Ô∏è‚É£ Merge Intervals

```python
def merge_intervals(intervals):
    if not intervals:
        return []
    
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]
    
    for current in intervals[1:]:
        last = merged[-1]
        if current[0] <= last[1]:
            last[1] = max(last[1], current[1])
        else:
            merged.append(current)
    
    return merged

# Edge Cases
print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))  # [[1,6],[8,10],[15,18]]
print(merge_intervals([[1,10],[2,3]]))  # [[1,10]]
print(merge_intervals([]))  # []
```

**Time Complexity:** O(n log n)
Handles nested and touching intervals.

### 6Ô∏è‚É£ Sliding Window ‚Äì Longest Ones With k Replacements

```python
def longest_ones(nums, k):
    left = 0
    max_len = 0
    zeros_count = 0
    
    for right in range(len(nums)):
        if nums[right] == 0:
            zeros_count += 1
        
        while zeros_count > k:
            if nums[left] == 0:
                zeros_count -= 1
            left += 1
        
        max_len = max(max_len, right - left + 1)
    
    return max_len

# Edge Cases
print(longest_ones([1,0,1,1,0], 1))  # 4
print(longest_ones([0,0,0], 0))      # 0
print(longest_ones([1,1,1], 2))      # 3
```

**Time Complexity:** O(n)
Handles all zeros, all ones, k=0.

### 7Ô∏è‚É£ Bit Manipulation ‚Äì Missing Number

```python
def missing_number(nums):
    n = len(nums)
    xor = 0
    for i in range(n+1):
        xor ^= i
    for num in nums:
        xor ^= num
    return xor

# Edge Cases
print(missing_number([3,0,1]))  # 2
print(missing_number([]))       # 0
print(missing_number([0]))      # 1
```

**Time Complexity:** O(n)
Handles empty array and 0/1 correctly.

### 8Ô∏è‚É£ Valid Parentheses

```python
def is_valid_parentheses(s):
    stack = []
    mapping = {')':'(', ']':'[', '}':'{'}
    
    for char in s:
        if char in mapping.values():
            stack.append(char)
        elif char in mapping:
            if not stack or stack.pop() != mapping[char]:
                return False
    return not stack

# Edge Cases
print(is_valid_parentheses("()[]{}"))  # True
print(is_valid_parentheses("([)]"))    # False
print(is_valid_parentheses(""))        # True
print(is_valid_parentheses("))"))      # False
```

**Time Complexity:** O(n)
Handles empty string and unbalanced parentheses.

### 9Ô∏è‚É£ Rotate Array

```python
def rotate_array(nums, k):
    n = len(nums)
    if n == 0:
        return nums
    
    k = k % n
    return nums[-k:] + nums[:-k]

# Edge Cases
print(rotate_array([1,2,3,4,5,6,7], 3))  # [5,6,7,1,2,3,4]
print(rotate_array([1], 10))             # [1]
print(rotate_array([], 3))               # []
```

**Time Complexity:** O(n)
Handles k > n, empty array, single element.

### üîü First Non-Repeating Character

```python
def first_unique_char(s):
    from collections import Counter
    freq = Counter(s)
    
    for char in s:
        if freq[char] == 1:
            return char
    return None

# Edge Cases
print(first_unique_char("aabbcdd"))  # 'c'
print(first_unique_char("aabb"))     # None
print(first_unique_char(""))         # None
```

**Time Complexity:** O(n)
Handles all repeating characters and empty string.
