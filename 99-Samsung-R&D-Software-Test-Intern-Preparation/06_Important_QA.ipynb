### üîπ 1. What is Software Testing?

**Answer:**
Software testing is the process of verifying and validating whether a software application meets the specified requirements and works correctly without defects. It ensures quality, reliability, security, and performance before release.

---

### üîπ 2. What is the difference between SDLC and STLC?

**Answer:**

| SDLC (Software Development Life Cycle)                    | STLC (Software Testing Life Cycle)                                                 |
| --------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| Defines phases of software development                    | Defines phases of testing                                                          |
| Includes requirement, design, coding, testing, deployment | Includes requirement analysis, test planning, test case design, execution, closure |
| Focuses on development                                    | Focuses on validation                                                              |

STLC is a part of SDLC.

### üîπ 4. What is a Test Case?

**Answer:**
A test case is a document that contains:

- Test Case ID
- Description
- Preconditions
- Test Steps
- Expected Result
- Actual Result
- Status (Pass/Fail)

It verifies a specific functionality.
