````python
# Remove trailing whitespace (rstrip)

favorite_language = 'python '
print(f"Original: '{favorite_language}'")

favorite_language = favorite_language.rstrip()
print(f"After rstrip: '{favorite_language}'")

print("--------------------")

# Remove leading whitespace (lstrip)

favorite_language = ' python'
print(f"Original: '{favorite_language}'")

favorite_language = favorite_language.lstrip()
print(f"After lstrip: '{favorite_language}'")

print("--------------------")

# Remove both leading and trailing whitespace (strip)

favorite_language = ' python '
print(f"Original: '{favorite_language}'")

favorite_language = favorite_language.strip()
print(f"After strip: '{favorite_language}'")

print("--------------------")

# Stripping other characters (not just spaces)

text = '...python...'
print(f"Original: '{text}'")

text = text.strip('.')
print(f"After strip('.'): '{text}'")
````

````python
# Removing a prefix

nostarch_url = 'https://nostarch.com'
print(f"Original: {nostarch_url}")

nostarch_url = nostarch_url.removeprefix('https://')
print(f"After removeprefix: {nostarch_url}")

print("--------------------")
# Removing a suffix

filename = 'example.txt'
print(f"Original: {filename}")

filename = filename.removesuffix('.txt')
print(f"After removesuffix: {filename}")

print("--------------------")
# Combining both 

url = 'https://www.python.org/'
print(f"Original: {url}")

url = url.removeprefix('https://').removesuffix('.org/')
print(f"Cleaned URL: {url}")
````

### def clean_url(url: str) -> str:

```python
def clean_url(url: str) -> str:
```

This line is a **function definition** in Python, and it uses **type hints** (also called _type annotations_). Here’s what each part means:

1. `def clean_url(...)`

- `def` starts the definition of a function.
- `clean_url` is the function’s name. You’ll call it later like `clean_url("https://example.com/")`.

2. `(url: str)`

- Inside the parentheses are the **parameters** the function accepts.
- `url` is the parameter name.
- `: str` is a **type hint** telling Python (and other developers) that `url` is expected to be a string.

Example:

```python
clean_url("https://nostarch.com")   # ✅ valid, it's a string
clean_url(12345)                    # ⚠️ not valid, type hint says it should be str
```

_(Note: Python won’t enforce this at runtime by default — it’s more for readability and tools like linters or IDEs. But libraries like `mypy` can check types.)_

3. `-> str`

- This is the **return type hint**.
- It means the function is expected to return a string.

Example:

```python
result = clean_url("https://python.org/")
print(type(result))   # <class 'str'>
```

Putting it all together
So the line:

```python
def clean_url(url: str) -> str:
```

means:  
**“Define a function called `clean_url` that takes one argument `url` (which should be a string) and will return a string.”**


````python
def clean_url(url: str) -> str:
    
    # Remove protocol prefix
    url = url.removeprefix("https://")
    url = url.removeprefix("http://")
    
    # Remove trailing slash if present
    url = url.removesuffix("/")
    
    return url


# Example usage
print(clean_url("https://nostarch.com"))       # Output: nostarch.com
print(clean_url("http://example.org/"))        # Output: example.org
print(clean_url("https://www.python.org/"))    # Output: www.python.org
````

````python
# Defining a large number with underscores for readability
universe_age = 14_000_000_000

# Printing the number
print(universe_age)

# Key Points:
# Underscores can be placed between digits in integers, floats, binary, hexadecimal, and octal numbers.
# Python ignores them when storing the value — they’re only for human readability.
# You can’t put underscores at the start or end of a number, or next to a decimal point.

# Large integer
population = 8_100_000_000
print(population)   # 8100000000

# Float with underscores
pi_approx = 3.141_592_653
print(pi_approx)    # 3.141592653

# Binary with underscores
binary_num = 0b1010_1111_0101
print(binary_num)   # 2805

# Hexadecimal with underscores
hex_num = 0xFF_FF_FF
print(hex_num)      # 16777215

````
