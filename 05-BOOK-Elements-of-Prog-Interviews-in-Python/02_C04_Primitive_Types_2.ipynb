## 4.7: Compute `x^y` # pg-32


### ðŸ”¹ Question (Interview Style)

**Problem:**  
Write a program that takes a double \(x\) and an integer \(y\), and returns \(x^y\) (that is, \(x\) raised to the power \(y\)).  
You can ignore overflow and underflow.

**Hint:**  
Exploit mathematical properties of exponentiation:

- \(x^y = (x^{y/2})^2\) if \(y\) is even
- \(x^y = x \cdot x^{y-1}\) if \(y\) is odd
- \(x^0 = 1\)

### ðŸ”¹ Explanation

The naive way is to multiply \(x\) by itself \(y\) times, but that takes \(O(y)\) time.  
Instead, we use **Exponentiation by Squaring** (fast power algorithm):

- Divide the problem into halves.
- If \(y\) is even, compute \(x^{y/2}\) once and square it.
- If \(y\) is odd, reduce it to \(x \cdot x^{y-1}\).
- This reduces the complexity to \(O(\log y)\).

This is the standard efficient way to compute powers in interviews.

### ðŸ”¹ Python Solution

```python
def power(x: float, y: int) -> float:
    if y == 0:
        return 1.0
    if y < 0:
        # Handle negative exponent: x^(-y) = 1 / (x^y)
        return 1.0 / power(x, -y)

    half = power(x, y // 2)

    if y % 2 == 0:
        return half * half
    else:
        return x * half * half

# Examples
print(power(2.0, 10))   # 1024.0
print(power(3.0, 5))    # 243.0
print(power(2.0, -3))   # 0.125
```

### ðŸ”¹ Step-by-Step Example: \(2^{10}\)

1. \(y = 10\) â†’ even â†’ compute \(2^5\).
2. \(y = 5\) â†’ odd â†’ compute \(2 \cdot 2^4\).
3. \(y = 4\) â†’ even â†’ compute \(2^2\).
4. \(y = 2\) â†’ even â†’ compute \(2^1\).
5. \(y = 1\) â†’ odd â†’ compute \(2 \cdot 2^0 = 2\).
6. Build back up:
   - \(2^2 = 4\)
   - \(2^4 = 16\)
   - \(2^5 = 32\)
   - \(2^{10} = 1024\)

### ðŸ”¹ Complexity

- **Time:** \(O(\log y)\)
- **Space:** \(O(\log y)\) (recursion depth)

âœ… So the problem is solved efficiently using **exponentiation by squaring**, which is the standard interview approach.

Would you like me to also show the **iterative version** (no recursion, pure loop) since some interviewers prefer that style for clarity and avoiding recursion overhead?


````python
def power(x: float, y: int) -> float:
    if y == 0:
        return 1.0
    if y < 0:
        # Handle negative exponent: x^(-y) = 1 / (x^y)
        return 1.0 / power(x, -y)
    
    half = power(x, y // 2)
    
    if y % 2 == 0:
        return half * half
    else:
        return x * half * half

# Examples
print(power(2.0, 10))   # 1024.0
print(power(3.0, 5))    # 243.0
print(power(2.0, -3))   # 0.125
````

Letâ€™s write the **iterative version of fast exponentiation (exponentiation by squaring)**. This avoids recursion and uses a simple loop, which many interviewers prefer for clarity and efficiency.

### ðŸ”¹ Iterative Fast Power Algorithm

We want to compute \(x^y\) where:

- \(x\) is a double (float in Python).
- \(y\) is an integer (can be positive, zero, or negative).

### Key idea:

- Use binary decomposition of \(y\).
- While \(y > 0\):
  - If the current bit of \(y\) is 1 â†’ multiply result by \(x\).
  - Square \(x\).
  - Shift \(y\) right (divide by 2).

This works in \(O(\log y)\) time.

### ðŸ”¹ Python Implementation (Iterative)

```python
def power_iterative(x: float, y: int) -> float:
    if y == 0:
        return 1.0

    # Handle negative exponent
    if y < 0:
        x = 1.0 / x
        y = -y

    result = 1.0
    while y > 0:
        if y & 1:       # if current bit of y is 1
            result *= x
        x *= x          # square the base
        y >>= 1         # shift exponent right (divide by 2)
    return result

# Examples
print(power_iterative(2.0, 10))   # 1024.0
print(power_iterative(3.0, 5))    # 243.0
print(power_iterative(2.0, -3))   # 0.125
```

### ðŸ”¹ Step-by-Step Example: \(2^{10}\)

- Start: result = 1, x = 2, y = 10 (`1010` in binary).
- Bit0 = 0 â†’ skip multiply, square x â†’ x = 4, y = 5.
- Bit1 = 1 â†’ result \*= 4 â†’ result = 4, square x â†’ x = 16, y = 2.
- Bit2 = 0 â†’ skip multiply, square x â†’ x = 256, y = 1.
- Bit3 = 1 â†’ result \*= 256 â†’ result = 1024, square x â†’ x = 65536, y = 0.
- Done â†’ result = 1024.

âœ… This iterative version is **efficient, avoids recursion overhead, and is interviewâ€‘friendly**.


````python
def power_iterative(x: float, y: int) -> float:
    if y == 0:
        return 1.0
    
    # Handle negative exponent
    if y < 0:
        x = 1.0 / x
        y = -y
    
    result = 1.0
    while y > 0:
        if y & 1:       # if current bit of y is 1
            result *= x
        x *= x          # square the base
        y >>= 1         # shift exponent right (divide by 2)
    return result

# Examples
print(power_iterative(2.0, 10))   # 1024.0
print(power_iterative(3.0, 5))    # 243.0
print(power_iterative(2.0, -3))   # 0.125
````

**Hint:**

There are two common approaches:

1. **String-based approach (easy):**
   - Convert the integer to a string.
   - Reverse the string.
   - Convert back to integer.
   - Handle negative numbers separately.

2. **Mathematical approach (bit more algorithmic):**
   - Extract digits using modulo (`% 10`).

### ðŸ”¹ Python Solutions

### âœ… String-based solution

```python
def reverse_digits_string(x: int) -> int:
    if x < 0:
        return -int(str(-x)[::-1])  # remove sign, reverse, reapply sign
    return int(str(x)[::-1])

# Examples

print(reverse_digits_string(-314))  # -413
```

### âœ… Mathematical solution

```python
def reverse_digits_math(x: int) -> int:
    sign = -1 if x < 0 else 1
    x = abs(x)

    while x > 0:
        digit = x % 10
        result = result * 10 + digit
        x //= 10

# Examples
print(reverse_digits_math(42))    # 24
print(reverse_digits_math(-314))  # -413
```

### ðŸ”¹ Complexity

- **Time:** \(O(n)\), where \(n\) = number of digits.

**Problem:**

A **palindrome** is a number (or string) that reads the same forwards and backwards.

**Examples (True):**

- `0`, `1`, `7`, `11`, `121`, `333`, `2147447412`

**Examples (False):**

### ðŸ”¹ Explanation

There are two main approaches:

1. **String-based approach (easy):**
   - Convert the integer to a string.
   - Check if the string equals its reverse.
   - Negative numbers are automatically not palindromes.

2. **Mathematical approach (algorithmic, interviewâ€‘friendly):**
   - Extract digits from both ends (most significant and least significant).

   - Stop when digits meet in the middle.

### ðŸ”¹ Python Solutions

### âœ… String-based solution

```python
def is_palindrome_string(x: int) -> bool:

print(is_palindrome_string(-121))  # False
print(is_palindrome_string(100))   # False
```

### âœ… Mathematical solution

```python
def is_palindrome_math(x: int) -> bool:
    if x < 0:
        return False

    # Find divisor to extract most significant digit
    divisor = 1
    while x // divisor >= 10:
        divisor *= 10

    while x:
        left = x // divisor      # most significant digit
        right = x % 10           # least significant digit

        if left != right:
            return False

        # Remove left and right digits
        x = (x % divisor) // 10
        divisor //= 100

    return True

print(is_palindrome_math(2147447412)) # True
print(is_palindrome_math(100))        # False
```

- Remove digits â†’ `x = (121 % 100) // 10 = 2`, divisor = 1.
- Compare left = `2 // 1 = 2`, right = `2 % 10 = 2` â†’ match.
- Done â†’ palindrome âœ….

### ðŸ”¹ Complexity

- **Math method:** \(O(n)\) time, \(O(1)\) space.

âœ… So the problem can be solved either by **string reversal** (quick) or by **digit comparison** (efficient and interviewâ€‘friendly).

## 4.10: Generate Uniform Random Numbers #pg-34


### ðŸ”¹ Question

**Problem:**  
Write a program that generates a random integer \(i\) between \(a\) and \(b\) (inclusive), given only access to a random number generator that produces `0` or `1` with equal probability.

All values in \([a, b]\) should be **equally likely**.

**Hint:**  
Think about how you would mimic a threeâ€‘sided coin using a twoâ€‘sided coin.

### ðŸ”¹ Explanation

We are given only a **fair coin flip generator** (`rand_bit()` â†’ returns 0 or 1 with equal probability).  
We need to extend this to generate numbers uniformly in any range \([a, b]\).

### Steps:

1. Compute the range size:  
   \[
   n = b - a + 1
   \]
   (number of possible outcomes).

2. Find the smallest power of 2 greater than or equal to \(n\).
   - Example: if \(n = 6\), smallest power of 2 â‰¥ 6 is 8.
   - This ensures we can generate enough distinct outcomes using bits.

3. Generate random bits to form a number in \([0, 2^k - 1]\).
   - If the number is less than \(n\), accept it.
   - If not, reject and try again (this ensures fairness).

4. Return \(a + \text{number}\).

This is called the **rejection sampling method**.

### ðŸ”¹ Python Solution

```python
import random

def rand_bit():
    return random.randint(0, 1)  # unbiased coin flip

def uniform_random(a, b):
    n = b - a + 1
    # Find smallest power of 2 >= n
    k = 0
    while (1 << k) < n:
        k += 1

    while True:
        # Generate k random bits
        num = 0
        for i in range(k):
            num = (num << 1) | rand_bit()

        if num < n:
            return a + num

# Examples
print(uniform_random(1, 6))   # Random number between 1 and 6
print(uniform_random(10, 15)) # Random number between 10 and 15
```

### ðŸ”¹ Example Walkthrough: Range [1, 6]

- \(n = 6\).
- Smallest power of 2 â‰¥ 6 is 8 â†’ need 3 bits.
- Generate 3 random bits â†’ number between 0 and 7.
- If number < 6 â†’ accept.
- Else (6 or 7) â†’ reject and retry.
- Result is uniformly distributed between 1 and 6.

### ðŸ”¹ Complexity

- **Time:** Expected \(O(1)\) (though rejection may cause retries).
- **Space:** \(O(1)\).

âœ… This method guarantees **fairness**: every integer in \([a, b]\) has equal probability, even though we only start with a binary coin flip generator.


````python
import random

def rand_bit():
    return random.randint(0, 1)  # unbiased coin flip

def uniform_random(a, b):
    n = b - a + 1
    # Find smallest power of 2 >= n
    k = 0
    while (1 << k) < n:
        k += 1
    
    while True:
        # Generate k random bits
        num = 0
        for i in range(k):
            num = (num << 1) | rand_bit()
        
        if num < n:
            return a + num

# Examples
print(uniform_random(1, 6))   # Random number between 1 and 6
print(uniform_random(10, 15)) # Random number between 10 and 15
````

### ðŸ”¹ Key Idea

- A dice roll has **6 equally likely outcomes**.
- We need to generate numbers uniformly in \([1, 6]\).
- Using rejection sampling:
  1. Generate enough random bits to cover at least 6 outcomes.
  2. The smallest power of 2 â‰¥ 6 is **8** â†’ so we need **3 bits**.

  4. If the number is in \([0, 5]\), accept it.
  5. If the number is 6 or 7, reject and retry.

### ðŸ”¹ Explanation

Rectangles aligned with the axes can be represented by their **bottomâ€‘left** and **topâ€‘right** corners:

- Rectangle A: \((x1, y1)\) bottomâ€‘left, \((x2, y2)\) topâ€‘right

### Intersection logic:

   - Left boundary = \(\max(x1, x3)\)

   - Overlap exists if `left < right`.

2. **Yâ€‘axis overlap:**
   - Bottom boundary = \(\max(y1, y3)\)
   - Top boundary = \(\min(y2, y4)\)
   - Overlap exists if `bottom < top`.

3. If both overlaps exist â†’ intersection rectangle = \((left, bottom)\) to \((right, top)\).
4. Otherwise â†’ no intersection.

### ðŸ”¹ Python Solution

```python
def rectangle_intersection(rectA, rectB):
    # rect = (x1, y1, x2, y2)
    x1, y1, x2, y2 = rectA
    x3, y3, x4, y4 = rectB

    # Compute overlap boundaries
    left   = max(x1, x3)
    right  = min(x2, x4)

    if left < right and bottom < top:

    else:
```
